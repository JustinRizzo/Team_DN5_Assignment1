## Investigate the impact of discount

### Subtask:
Analyze how the `discount` column affects `profit` and `profit_margin`. This could involve grouping data by discount levels and comparing profitability metrics.


**Reasoning**:
Construct and execute a SQL query to analyze the relationship between discount and profitability metrics, then display the results.



```python
sql = """
SELECT
    discount,
    AVG(profit) AS average_profit,
    AVG(profit_margin) AS average_profit_margin
FROM
    `myproj5900.lab1_eda.lab1_eda`
GROUP BY
    discount
ORDER BY
    discount
"""
df_discount_profit = client.query(sql).to_dataframe()
display(df_discount_profit)
```

## Explore the impact of category and subcategory

### Subtask:
Explore the impact of category and subcategory on product profitability. Analyze if certain category or subcategory have higher or lower average profit margins.


**Reasoning**:
Write a SQL query to calculate the average profit and average profit margin for each category, execute the query, load the results into a DataFrame, and display it sorted by average profit margin in descending order.



```python
sql = """
SELECT category,
       AVG(profit) AS average_profit,
       AVG(profit_margin) AS average_profit_margin
FROM `myproj5900.lab1_eda.lab1_eda`
GROUP BY category
ORDER BY average_profit_margin DESC
"""
df_category_profit = client.query(sql).to_dataframe()
display(df_category_profit)
```

**Reasoning**:
Write a SQL query to calculate the average profit and average profit margin for each subcategory, execute the query, load the results into a new DataFrame, and display it sorted by average profit margin in descending order.



```python
sql = """
SELECT subcategory,
       AVG(profit) AS average_profit,
       AVG(profit_margin) AS average_profit_margin
FROM `myproj5900.lab1_eda.lab1_eda`
GROUP BY subcategory
ORDER BY average_profit_margin DESC
"""
df_subcategory_profit = client.query(sql).to_dataframe()
display(df_subcategory_profit)
```

## Examine the effect of region

### Subtask:
Investigate if `region` plays a role in product profitability.


**Reasoning**:
Calculate the average profit and average profit margin, grouped by region, and display the results.



```python
sql = """
SELECT region,
       AVG(profit) AS average_profit,
       AVG(profit_margin) AS average_profit_margin
FROM `myproj5900.lab1_eda.lab1_eda`
GROUP BY region
ORDER BY average_profit_margin DESC
"""
df_region_profit = client.query(sql).to_dataframe()
display(df_region_profit)
```

## Look into the influence of customer and manufactory

### Subtask:
Analyze if certain `customer` or `manufactory` are associated with higher or lower profitability.


**Reasoning**:
Calculate the average profit and average profit margin for each customer and manufactory to identify if certain ones are associated with higher or lower profitability, then display the results.



```python
sql_customer = """
SELECT customer,
       AVG(profit) AS average_profit,
       AVG(profit_margin) AS average_profit_margin
FROM `myproj5900.lab1_eda.lab1_eda`
GROUP BY customer
ORDER BY average_profit_margin DESC
"""
df_customer_profit = client.query(sql_customer).to_dataframe()
display(df_customer_profit)

sql_manufactory = """
SELECT manufactory,
       AVG(profit) AS average_profit,
       AVG(profit_margin) AS average_profit_margin
FROM `myproj5900.lab1_eda.lab1_eda`
GROUP BY manufactory
ORDER BY average_profit_margin DESC
"""
df_manufactory_profit = client.query(sql_manufactory).to_dataframe()
display(df_manufactory_profit)
```

## Visualize findings

### Subtask:
Create visualizations to illustrate the relationships discovered in the analysis (e.g., scatter plots of discount vs. profit margin, bar charts of average profit margin by category).


**Reasoning**:
Import the necessary libraries for plotting and create the specified visualizations to show the relationships between profitability metrics and the influencing factors.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot of Average Profit Margin vs. Discount
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_discount_profit, x='discount', y='average_profit_margin')
plt.title('Average Profit Margin vs. Discount')
plt.show()

# Bar chart of Average Profit Margin by Category
plt.figure(figsize=(8, 5))
sns.barplot(data=df_category_profit, x='category', y='average_profit_margin', order=df_category_profit.sort_values('average_profit_margin', ascending=False)['category'])
plt.title('Average Profit Margin by Category')
plt.show()

# Bar chart of Top 10 Subcategories by Average Profit Margin
df_subcategory_profit_top10 = df_subcategory_profit.nlargest(10, 'average_profit_margin')
plt.figure(figsize=(12, 6))
sns.barplot(data=df_subcategory_profit_top10, x='subcategory', y='average_profit_margin', order=df_subcategory_profit_top10['subcategory'])
plt.title('Top 10 Subcategories by Average Profit Margin')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Bar chart of Average Profit Margin by Region
plt.figure(figsize=(8, 5))
sns.barplot(data=df_region_profit, x='region', y='average_profit_margin', order=df_region_profit.sort_values('average_profit_margin', ascending=False)['region'])
plt.title('Average Profit Margin by Region')
plt.show()
```

## Summary:

### Data Analysis Key Findings

*   A strong positive correlation was observed between total profit and total sales (0.92), and a moderate positive correlation between total profit and average profit margin (0.72).
*   Discount has a significant negative impact on profitability; as discount percentage increases, average profit and average profit margin decrease, becoming negative at higher discount levels.
*   The correlation between quantity and profit is weak positive (0.066), and the correlation between quantity and profit margin is very weak negative (-0.005).
*   Profitability varies significantly by category and subcategory. 'Technology' is the most profitable category on average, while subcategories like 'Labels', 'Paper', and 'Envelopes' show high average profit margins. Subcategories such as 'Tables', 'Bookcases', and 'Binders' have negative average profit margins.
*   Region plays a role in profitability, with the West region having the highest average profit margin (0.219) and the Central region having a negative average profit margin (-0.104).
*   Profitability varies across individual customers and manufactories, indicating that some are associated with higher or lower average profit margins.

### Insights or Next Steps

*   Focus on reducing discounts, especially for products in categories and subcategories with already low or negative profit margins, and in the Central region.
*   Develop targeted sales strategies to promote high-profitability categories and subcategories, potentially focusing on customers and regions that exhibit higher average profit margins.
